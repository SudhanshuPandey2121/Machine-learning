Seaborn

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

Note: Seaborn has some builtin datasets


```python
#total bill vs tip dataset
tips=sns.load_dataset('tips')  #this is will import it as a pandas dataset
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualise data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='sex',style= 'smoker',size='size')
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot to cluster different species

```python
#clustering it
#hue means on what basis data will be separated
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_width',y='petal_width',hue='species',data=iris)
```

```python
#loading titanic dataset
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.scatterplot(data=titanic,x='survived',y='pclass',hue='who')
```

count plot


```python
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',hue='pclass',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

Bar chart

```python
sns.barplot(x='sex',y='survived',hue='pclass',data=titanic)
```

```python
boston=pd.read_csv('/content/BostonHousing.csv')
```

```python
boston.head()
```

Distribution plot

```python
sns.distplot(boston['medv'])
```

Correlation matrix

```python
correlation=boston.corr()
```

```python
#constructing a heat map
#firstly constructing a figure space using matplotlib
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
