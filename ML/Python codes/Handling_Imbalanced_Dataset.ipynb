Imbalanced Dataset  
A dataset with an unequal class distribution  
for example:- diabetes dataset with 99 non diabetic and 1 diabetic data

```python
import numpy as np
import pandas as pd
```

```python
credit_card = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card.head()
```

Class --> 1 = fraudulent Transaction   
Class --> 0 = legit transaction

```python
credit_card['Class'].value_counts()
```

```python
credit_card.shape
```

This is highly imbalanced dataset

```python
#seperating legit and fraud transactions
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```


**Under-Sampling**  
buiding a dataset with a similar distribution of legit and fraud transactions


```python
legit_sample = legit.sample(n=25)
```

```python
legit_sample.shape
```

concatenate legit and fraud

```python
new_data=  pd.concat([legit_sample,fraud],axis=0)
#axis =0 means that we want to concatenate rowwise not columnwise
```
