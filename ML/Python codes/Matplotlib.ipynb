Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
y=np.sin(x)
```

```python
z=np.cos(x)
```

```python
plt.plot(x,y,'r-')
plt.plot(x,z,'g+')
plt.xlabel('Angle')
plt.ylabel('Value')
plt.title('trignometric ratios')
plt.show
```

```python
#parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','Spanish','Sanskrit','Hindi','Urdu']
speakers=[100,120,23,200,150]
ax.bar(languages,speakers)
plt.xlabel('languages')
plt.ylabel('Person')
plt.title('Speakers')
plt.show()
```

pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','Spanish','Sanskrit','Hindi','Urdu']
speakers=[100,120,23,200,150]
ax.pie(speakers,labels=languages,autopct='%1.1f%%')

plt.title('Speakers')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,20)
y=np.sin(x)
plt.scatter(x,y,color='g')
plt.show()
```

3D scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
